### 🔍 Tipos de Aprendizaje

| Tipo | Descripción | Ejemplos |
|:---|:---|:---|
| **Supervisado** | Datos etiquetados, aprender a predecir etiquetas | Regresión, Clasificación |
| **No Supervisado** | Datos no etiquetados, encontrar estructuras ocultas | Clustering, Reducción de dimensionalidad |
| **Por Refuerzo** | Agente aprende por prueba y error, maximizando recompensas | Juegos, Robótica, Control |

---

#### 📊 Aprendizaje Supervisado
```
Entrada → Modelo → Salida (etiqueta)
    ↑______________________|
         Retroalimentación (error)
```

#### 🔬 Aprendizaje No Supervisado
```
Datos sin etiquetas → Modelo → Grupos/Patrones encontrados
```

#### 🎮 Aprendizaje por Refuerzo
```
Estado → Agente → Acción → Recompensa → Nuevo Estado
```

**Fórmula de retorno:**  
$R_t = \sum_{i=t}^{\infty} \gamma^i r_i = \gamma^t r_t + \gamma^{t+1} r_{t+1} + ... + \gamma^{t+n} r_{t+n} + ...$  
donde $\gamma$ (gamma) es el factor de descuento: $0 < \gamma < 1$

## 📋 Ciclo de Vida de un Proyecto de ML

<img src="https://miro.medium.com/v2/resize:fit:1400/1*NcRoVUEXoim3k28dQI8dxg.png" width="700">

1. **Definición del problema**  
   *   ¿Qué queremos predecir/clasificar?
   *   ¿Cuál es la métrica de éxito?

2. **Recopilación de datos**  
   *   Fuentes: repositorios públicos, bases de datos, scraping, sensores
   *   🔗 [UC Irvine Machine Learning Repository](https://archive.ics.uci.edu/)
   *   🔗 [Kaggle Datasets](https://www.kaggle.com/datasets)
   *   🔗 [AWS Open Data](https://registry.opendata.aws/)
   *   🔗 [Data Portals](http://dataportals.org/)
   *   🔗 [Open Data Monitor](http://opendatamonitor.eu/)
   *   🔗 [Quandl](https://www.quandl.com/)

3. **Exploración y visualización de datos**  
   *   Entender distribuciones, correlaciones, patrones
   *   Identificar problemas (valores faltantes, outliers)

4. **Preprocesamiento y limpieza**  
   *   "Garbage In, Garbage Out" - Datos pobres → malas predicciones
   *   Manejo de valores faltantes
   *   Codificación de variables categóricas
   *   Normalización / Estandarización
   *   Reducción de dimensionalidad
   *   Eliminación de sesgos

5. **Selección y entrenamiento de modelos**  
   *   División train/test/validation
   *   Entrenamiento
   *   Validación cruzada

6. **Evaluación y ajuste fino**  
   *   Métricas de evaluación
   *   Ajuste de hiperparámetros
   *   Prevención de overfitting/underfitting

7. **Despliegue y monitoreo**  
   *   Puesta en producción
   *   Monitoreo de performance en el tiempo
   *   Re-entrenamiento periódico

### ⚠️ Importancia del Preprocesamiento

**"Entrada de basura → Salida de basura"**  
**"Datos pobres → Malas predicciones / modelado"**

El preprocesamiento es obligatorio para algunos modelos y ayuda a:
*   Disminuir sesgos en los datos
*   Identificar datos redundantes e irrelevantes
*   Reducir la dimensionalidad
*   Mejorar la convergencia del entrenamiento
*   Hacer los modelos más robustos
