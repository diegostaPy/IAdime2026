# Regresión lineal utilizando la equacion normal(método directo-pseudoinversa)

$\mu$

$\phi(x_{i})=w_{1} x+w_{0}$

$F(w)=\Sigma_{i=1}^{3}(y_{i}-\phi_{i})^{2}$

$F(w)=(y_{1}-\phi(x_{1})^2+(y_{2}-\phi(x_{2})^2+(y_{3}-\phi(x_{3})^2$

##Formulacion del problema

$P_{1}=(x_{1},y_{1})$

$P_{2}=(x_{2},y_{2})$

$P_{3}=(x_{3},y_{3})$

$y=\phi(x)=w_{1} x+w_{0}$

$F(w)=\Sigma_{i=1}^{3}Error_{i}^{2}$

$F(w)=(y_{1}-\phi(x_{1})^2+(y_{2}-\phi(x_{2})^2+(y_{3}-\phi(x_{3})^2$

$F(w)=(y_{1}-w_{1}x_{1}-w_{0})^2$+$(y_{2}-w_{1}x_{2}-w_{0})^2$
    +$(y_{3}-w_{1}x_{3}-w_{0})^2$
    
$\frac{dF}{dw_{1}}$=$-2x_{1}(y_{1}-w_{1}x_{1}-w_{0})$-
                 $2x_{2}(y_{2}-(w_{1}x_{2}-w_{0})$-
                 $2x_{3}(y_{3}-w_{1}x_{3}-w_{0})$

  

$\frac{dF}{dw_{0}}$=$-2(y_{1}-w_{1}x_{1}-w_{0})$-
    $2(y_{2}-w_{1}x_{2}-w_{0})$-
        $2(y_{3}-w_{1}x_{3}-w_{0})$


    


Igualando a cero 


$\frac{dF}{dw_{0}}$=0=$(x_{1}+x_{2}+x_{3})w_{1}+3w_{0}=(y_{1}+y_{2}+y_{3})$

$\frac{dF}{dw_{1}}$=0=$(x_{1}x_{1}+x_{2}x_{2}+x_{3}x_{3})w_{1}+(x_{1}+x_{2}+x_{3})w_{0}=(x_{1}y_{1}+x_{2}y_{2}+x_{3}y_{3})$



Que se puede reecribir como :
    $A^TA w=A^Tb$

$donde\\

A=\begin{bmatrix}
1&x_{1}\\
1&x_{2}\\
1&x_{3}\\
\end{bmatrix} $

$w=\begin{bmatrix}
w_{0}\\
w_{1}\\
\end{bmatrix} $



$b=\begin{bmatrix}
y_{1}\\
y_{2}\\
y_{b}\\
\end{bmatrix} $

# Fomulacion matricial
minimizar $F(w)=(Aw-b)(Aw-b)^T$ es equivalente a minimizar $\Sigma_{i=1}^{3}Error_{i}^{2}$

#calculando el gradiente $\nabla F=$
$2(Aw-b)^T A=0$

$A^TAw=A^Tb$

$w=(A^{T}A)^{-1}A^{T}b$

```python
from mpl_toolkits.mplot3d import Axes3D

# Datos de entrada
x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([2, 3, 4, 5, 6])
y = np.array([5, 6, 7, 8, 9])

# Construir la matriz de diseño ampliada
X = np.vstack([x1, x2, np.ones(len(x1))]).T

# Calcular la pseudoinversa
A_pseudo_inv = np.linalg.pinv(X)

# Calcular los parámetros de la regresión
params = np.dot(A_pseudo_inv, y)

# Graficar
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos de datos
ax.scatter(x1, x2, y, color='r', marker='o', label='Datos')

# Definir el plano de regresión
x1_range = np.linspace(min(x1), max(x1), 100)
x2_range = np.linspace(min(x2), max(x2), 100)
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)
y_pred = params[0]*x1_mesh + params[1]*x2_mesh + params[2]

# Graficar el plano de regresión
ax.plot_surface(x1_mesh, x2_mesh, y_pred, alpha=0.5, color='b', label='Regresión')

# Etiquetas y título
ax.set_xlabel('Variable Independiente 1')
ax.set_ylabel('Variable Independiente 2')
ax.set_zlabel('Variable Dependiente')
ax.set_title('Regresión Lineal Multivariada')


# Mostrar gráfico
plt.show()
```

### Ecuaciones de Regresión Lineal Ponderada

En la regresión lineal ponderada, cada punto de datos tiene un peso diferente según algún criterio, como su importancia o confiabilidad. El método de mínimos cuadrados ponderados se utiliza comúnmente para implementar la regresión lineal ponderada.

Para un conjunto de datos con $n$ observaciones, el modelo puede representarse como:

$$
y = \beta_1 x + \beta_0 + \epsilon
$$

Donde:
- $y$ es la variable dependiente.
- $x$ es la variable independiente.
- $\beta_1$ es el coeficiente de pendiente.
- $\beta_0$ es la intersección.
- $\epsilon$ representa el término de error.

La solución de mínimos cuadrados ponderados se puede calcular utilizando la siguiente fórmula:

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{W} \mathbf{X})^{-1} \mathbf{X}^T \mathbf{W} \mathbf{y}
$$

Donde:
- $\mathbf{X}$ es la matriz de diseño, incluyendo una columna de unos para el término de intersección.
- $\mathbf{W}$ es una matriz diagonal de pesos, donde cada peso corresponde a un punto de datos.
- $\mathbf{y}$ es el vector de valores observados.
- $\boldsymbol{\beta}$ es el vector de coeficientes ($\beta_1$ y $\beta_0$).

Una vez que se obtienen los coeficientes, la línea de regresión lineal ponderada puede expresarse como:

$$
\hat{y} = \hat{\beta}_1 x + \hat{\beta}_0
$$

Donde $\hat{y}$ representa los valores predichos.


Los pesos en la práctica en datos reales pueden representar la importancia o confiabilidad de cada punto de datos en el proceso de regresión. Aquí hay algunos ejemplos de cómo se pueden utilizar los pesos en datos reales:

    Fiabilidad de los datos: En algunos casos, ciertos puntos de datos pueden ser más confiables que otros debido a la forma en que se recopilaron o a su precisión. Por ejemplo, en mediciones físicas, algunas mediciones pueden tener una precisión mayor que otras debido a la calidad del equipo utilizado o a la experiencia del observador. Los pesos pueden utilizarse para dar más importancia a los puntos de datos más precisos.

    Importancia relativa: En conjuntos de datos desequilibrados o sesgados, donde algunos subconjuntos de datos son más importantes que otros, los pesos pueden utilizarse para reflejar la importancia relativa de los diferentes subconjuntos de datos. Por ejemplo, en problemas de clasificación binaria donde una clase es mucho más rara que la otra, los datos de la clase minoritaria pueden tener un peso mayor para compensar el desequilibrio.

    Corrección de sesgo: En algunos casos, los datos pueden estar sesgados o no representativos de la población subyacente. Los pesos pueden utilizarse para corregir este sesgo dando más peso a los puntos de datos que son más representativos de la población objetivo.

    Variabilidad de los datos: En series temporales u otros conjuntos de datos donde la varianza de los errores no es constante, los pesos pueden utilizarse para tener en cuenta la variabilidad heterocedástica. En tales casos, los pesos pueden ser inversamente proporcionales a la varianza de los errores, lo que significa que los puntos de datos con menor variabilidad tendrán un peso mayor.

```python

X = np.random.rand(100, 1) 
y = 2 * X.squeeze() + 1 + np.random.normal(scale=0.5, size=100)  

pesos= 1/y#np.random.rand(100)  

A = np.hstack((X, np.ones_like(X))) 

W = np.diag(pesos)  
ajuste = np.linalg.inv(A.T.dot(W).dot(A)).dot(A.T).dot(W.dot(y)) #w
#np.linalg.inv(A.T @ W @ A) @ A.T @ (W @ y)
print(ajuste)
```
